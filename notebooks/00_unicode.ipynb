## 2.2 Unicode encodings


```python

# this encodes a string as utf-8


utf_8 = "abcd".encode("utf-8")
```

```python
utf_16 = "abcd".encode("utf-16")

utf_8 == utf_16
```

```python
test_string = "hello! こんにちは!"

test_string.encode("utf-8") == test_string.encode("utf-16")
```

:( not the same bro

```python
# Bytes in utf-8

print(f"UTF-8 representation of {test_string}")
for i, b in enumerate(test_string.encode("utf-8")):
    print(f"{i:10} -> {b}")

print(f"UTF-16 representation of {test_string}")
for i, b in enumerate(test_string.encode("utf-16")):
    print(f"{i:10} -> {b}")
```

```python
# Bytes in utf-8
test_string = "こんにちは!"
print(f"UTF-8 representation of {test_string}")
for i, b in enumerate(test_string.encode("utf-8")):
    print(f"{i:10} -> {b}")

print(f"UTF-16 representation of {test_string}")
for i, b in enumerate(test_string.encode("utf-16")):
    print(f"{i:10} -> {b}")
```

UTF-8: ASCII-compatible. Would likely be smaller for English/Internet text.
UTF-16: More compact in Chinese and in alphabets distinct from Latin
UTF-32: Fixed width. Not 

```python
test_string = "hello! \u1020"

print(test_string)
encoded = test_string.encode("utf-8")
print(encoded)
bytewise_decoding = [bytes([b]).decode("utf-8") for b in encoded]

print("".join(bytewise_decoding))
```
